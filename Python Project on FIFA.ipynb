```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
countries=fifa['nationality']
countries.value_counts()
```

```python
#top 10 nationalities
countries.value_counts()[0:10]
```

```python
#Bar Plot
plt.figure(figsize=(15,10))
plt.bar(list(countries.value_counts()[0:10].keys()),list(countries.value_counts()[0:10]),color='red')
plt.xlabel('Nationality')
plt.ylabel('Number of players')
plt.show()
```

```python
# here we can observe that, England has most number of players than other countries
```

```python
salary=fifa[['short_name','value_eur','wage_eur']]
salary.head()
```

```python
#Top 5 most valuable & highest paid player
salary=salary.sort_values(by=['value_eur','wage_eur'],ascending=False)
salary.head()
```

```python
#here we can see that Neymar is most valuable player, and Messi is highest paid player
```

```python
#Bar Plot
plt.figure(figsize=(17,7))
plt.bar(list(salary['short_name'])[0:10],list(salary['wage_eur'])[0:10],color='blue')
plt.xlabel('Name of PLayer')
plt.ylabel('Salary in Euros')
plt.show()
```

```python
#Bar Plot
plt.figure(figsize=(17,7))
plt.bar(list(salary['short_name'])[0:10],list(salary['value_eur'])[0:10],color='violet')
plt.xlabel('Name of PLayer')
plt.ylabel('Value in Euros')
plt.show()
```
